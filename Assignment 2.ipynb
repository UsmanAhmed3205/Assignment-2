# Scenario 1: Grocery Shopping List
### You are creating a program to manage a grocery shopping list. Users should be able to add items,remove items, and display the current list.

```python
print('''=======================
 Grocery Shopping List 
=======================''')
shoppingList = ["Egg" , "Rice" , "Milk" , "Oil"]
while True:
    print('''
---------Menu----------
| 1. Add an item      |
| 2. Remove an item   |
| 3. Display the List |
| 4. Exit             |
-----------------------''')
    userInput = int(input("Choose 1, 2 , 3 or 4: "))
    
    
    if userInput == 1:
        tempVar="a"
        while tempVar == "a":
            addItem = input('''Enter item to add :
(enter a space to finish the input)''')
            if addItem == "":
                break
            else:
                shoppingList.append(addItem)


    elif userInput == 2:
        removeItem = int(input("Enter item number to be removed: "))
        if removeItem <= len(shoppingList)+1 and removeItem >= 0 :
            shoppingList.pop(removeItem-1)
            
        else:
            print("Incorrect item number")
        
    elif userInput == 3:
        for i in range(len(shoppingList)):
            print(i+1 , shoppingList[i])
            
            
    elif userInput == 4:
        print("Thank You for Shopping!")
        break
```

# Scenario 4: Password Strength Checker

#### You are creating a program to check the strength of passwords based on certain criteria.
#### • Password should be at least 8 characters long.
#### • Password should contain at least one digit.
#### • Password should contain at least one letter.

```python
print('''===========================
 Password Strength Checker
===========================''')

userinput = input("Enter Password: ")
passlen = len(userinput)
pole = 0

if passlen >= 8:
    print("✔️ Password is 8 characters long ")
    pole += 1
else :
    print("Password should be at least 8 characters long.")
    
tempdigit = 0
for i in userinput:
    if i.isdigit():
        tempdigit = 1
        
if tempdigit == 1:
    print("✔️ Password contains atleast one digit.")
    pole += 1
else:
    print("Password should contain at least one digit.")
    
templetter = 0
for i in userinput:
    if i.isalpha():
        templetter = 1
        
if templetter == 1:
    print("✔️ Password contains atleast one letter.")
    pole += 1
else:
    print("Password should contain at least one letter.")
        
if pole == 1:
    print("Password strength is too Weak")
if pole == 2:
    print("Password strength is Weak")    
if pole == 3:
    print("Password strength is Strong")    
    
```

# Scenario 05: Voting System (Annual Employee Recognition Awards)
#### You are developing a simple voting system for a contest. Users can vote for their Favorite option, and you need to count the votes.

```python
candidatesList = ["Candidate A", "Candidate B", "Candidate C"]
Avotes = 0
Bvotes = 0
Cvotes = 0
while True:  
    print('''
====================
List of Candidates
====================''')
    for i in range(len(candidatesList)):
        print(i+1 , candidatesList[i])
  
    userInput = int(input("Enter the Candidate Number to Vote: "))

    if userInput == 1:
        print('''You Voted for Candidate A
Thank You for Voting!''')
        Avotes += 1
    elif userInput == 2:
        print('''You Voted for Candidate A
Thank You for Voting!''')
        Bvotes += 1
    elif userInput == 3:
        print('''You Voted for Candidate A
Thank You for Voting!''')
        Cvotes += 1
    elif userInput == 0:
        break
    else:
        print("Enter correct Candidate number or zero to exit")

print(f"===========\n   RESULT   \n===========\nvotes for Candidate A = {Avotes} \n votes for Candidate B = {Bvotes} \nvotes for Candidate C = {Cvotes}")
```

 # Scenario 2: Student Grades
#### You are managing student grades using a dictionary. You need to calculate the average grade. Take atleast 5 Students grades & then calculate the average.

# Scenario 3: Word Frequency Counter
#### You are given a list of words, and you need to count the frequency of each word.
#### word_list = ["apple", "banana", "apple", "orange", "banana", "grape", "apple"]

```python
word_list = ["apple", "banana", "apple", "orange", "banana", "grape", "apple"]

word_count = {}

for word in word_list:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(f"{word}: {count}")
```
